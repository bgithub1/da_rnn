```python
import torch 
from torch import nn
from torch.autograd import Variable
from torch import optim
import torch.nn.functional as F

def pred_model(m_train_size,m_batch_size,m_x_shape_0,m_x,m_y,m_t):

    y_pred = np.zeros(m_x_shape_0 - m_train_size)
    on_train=False
    rlp = range(len(y_pred))
    b_idx = np.array(rlp)

    i=0
    while i < len(y_pred):
        batch_idx = b_idx[i : (i + m_batch_size)]
        # calculate batch size, b/c on the last batch, it will be less that model.batch_size
        batch_length = len(batch_idx)
        X = np.zeros((batch_length, m_t - 1, m_x.shape[1]))
        y_history = np.zeros((batch_length, m_t - 1))
        for j in range(batch_length):
            if on_train:
                X[j, :, :] = m_x[range(batch_idx[j], batch_idx[j] + m_t - 1), :]
                y_history[j, :] = m_y[range(batch_idx[j],  batch_idx[j]+ m_t - 1)]
            else:
                test_index_start = batch_idx[j] + m_train_size - m_t
                test_index_end = batch_idx[j] + m_train_size - 1            
                X[j, :, :] = m_x[range(test_index_start, test_index_end), :]
                y_history[j, :] = m_y[range(test_index_start,  test_index_end)]

        y_history = Variable(torch.from_numpy(y_history).type(torch.FloatTensor))
        _, input_encoded = model.encoder(Variable(torch.from_numpy(X).type(torch.FloatTensor)))
        y_pred[i:(i + m_batch_size)] = model.decoder(input_encoded, y_history).cpu().data.numpy()[:, 0]

        i += m_batch_size
#     print(test_index_start,test_index_end,i,j,batch_idx)
    return y_pred

```

```python
m_train_size = model.train_size
m_batch_size = model.batch_size
m_x_shape_0 = model.X.shape[0]
m_x = model.X
m_y = model.y
m_t = model.T

yp_bp = pred_model(m_train_size,m_batch_size,m_x_shape_0,m_x,m_y,m_t)
yp_model = model.predict(on_train = False)
print(sum(yp_bp - yp_model))
```

```python
def pred_df(df_test):
    
    x_values = df_test[model.x_columns].as_matrix()
    m_y = df_test[model.ticker].as_matrix() - model.y_train_mean
    m_t = model.T
    m_batch_size = model.batch_size
    y_pred = np.zeros(x_values.shape[0] - m_t)
    on_train=False
    rlp = range(len(y_pred))
    b_idx = np.array(rlp)

    i=0
    while i < len(y_pred):
        batch_idx = b_idx[i : (i + m_batch_size)]
        # calculate batch size, b/c on the last batch, it will be less that model.batch_size
        batch_length = len(batch_idx)
        X = np.zeros((batch_length, m_t - 1, x_values.shape[1]))
        y_history = np.zeros((batch_length, m_t - 1))
        for j in range(batch_length):
            test_index_start = batch_idx[j]
            test_index_end = min(batch_idx[j],m_y.shape[0]) +  m_t - 1 
#             if test_index_start>=3000 or test_index_end>3000:
#                 return (test_index_start,test_index_end,i,j,batch_idx)
            X[j, :, :] = x_values[range(test_index_start, test_index_end), :]
            y_history[j, :] = m_y[range(test_index_start,  test_index_end)]

        y_history = Variable(torch.from_numpy(y_history).type(torch.FloatTensor))
        _, input_encoded = model.encoder(Variable(torch.from_numpy(X).type(torch.FloatTensor)))
        y_pred[i:(i + m_batch_size)] = model.decoder(input_encoded, y_history).cpu().data.numpy()[:, 0]

        i += m_batch_size
    return y_pred
```
