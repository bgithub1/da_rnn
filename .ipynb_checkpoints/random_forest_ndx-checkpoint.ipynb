NDX itself.
Use a rolling window of each component, so that each row represents a number of previous prices for each component.


Define an rsme function, and a function that compares rsme's and model.scores (R^2's) between predicted y values and actual y values.

```python
# def rmse(x,y): return math.sqrt(((x-y)**2).mean())
def rmse(x,y): return math.sqrt(((np.log(x)-np.log(y))**2).mean())

def print_score(mod,xt,yt,xv,yv):
    res = [rmse(mod.predict(xt), yt), rmse(mod.predict(xv), yv),
                mod.score(xt, yt), mod.score(xv, yv)]
    labels = ['rsme train','rsme test','R2 train','R2 test']
    if hasattr(mod, 'oob_score_'): 
        res.append(mod.oob_score_)
        labels.append('oob_score')
    df = pd.DataFrame({'function':labels,'values':res})
    print(df)
```

Create a function that creates "lag" values for any component stock of NDX

```python
def table2lags(table, max_lag, min_lag=0, separator='_'):
    """ Given a dataframe, return a dataframe with different lags of all its columns """
    values=[]
    for i in range(min_lag, max_lag + 1):
        values.append(table.shift(i).copy())
        values[-1].columns = [c + separator + str(i) for c in table.columns]
    return pd.concat(values, axis=1)
```

```python
#example usage of table2flags
df = pd.DataFrame({'y1':[1,2,3,4,5], 'y2':[10,20,40,50,30], 'z': [1,4,9,16,25]})
x = table2lags(df[['y1', 'y2']], 2)
print(type(x),x)
```

Examle 1:  relate AAPL and MSFT to NDS

```python
df = pd.read_csv('data/nasdaq100_padding.csv')
df.columns.values
```

```python
len(df)
```

Create a dataframe with just aapl and msft columns, and then 60 previous days prices for each.
Then, create a training set and test set of rows, where n = the number of test rows from the 
full DataFrame df.

```python
num_test_rows = 20000
num_validation_rows = 2000 # number of test rows
days_back = 30 # number of days of history to analyze for each component stock
df_sub = df.drop('NDX',axis=1).copy() # create a dataframe without NDX, which is the y vector
df_sub = df[['AAPL','MSFT']] # just juse aapl and msft
df_sub = table2lags(df_sub,days_back) # create the history columns
df_sub = df_sub.drop(['AAPL_0','MSFT_0'],axis=1)
X = df_sub.as_matrix() # make matrix of X values
y = df['NDX'].as_matrix() # make matrix of y
X = X[-num_test_rows:]
y = y[-num_test_rows:]
X_train = X[:-num_validation_rows].copy() # create a training set for X
X_test = X[-num_validation_rows:].copy()  # create a test set for X
y_train = y[:-num_validation_rows].copy() # same as above for y
y_test = y[-num_validation_rows:].copy() # same as above for y
m2 = rfr(n_estimators=10,n_jobs=-1,oob_score=True) # create model
%time m2.fit(X_train,y_train) # train the model
print_score(m2,X_train,y_train,X_test,y_test) # see how well you did
```

example using dataframe for X rather than a matrix

```python
num_test_rows = 8000
num_validation_rows = 200 # number of test rows
days_back = 30 # number of days of history to analyze for each component stock
df_sub = df.drop('NDX',axis=1).copy() # create a dataframe without NDX, which is the y vector
df_sub = df[['AAPL','MSFT']] # just juse aapl and msft
df_sub = table2lags(df_sub,days_back) # create the history columns
df_sub = df_sub.drop(['AAPL_0','MSFT_0'],axis=1)
X = df_sub.copy()
y = df['NDX']
X = X.iloc[-num_test_rows:]
y = y.iloc[-num_test_rows:]
X_train = X.iloc[:-num_validation_rows].copy() # create a training set for X
X_test = X.iloc[-num_validation_rows:].copy()  # create a test set for X
y_train = y.iloc[:-num_validation_rows].copy() # same as above for y
y_test = y.iloc[-num_validation_rows:].copy() # same as above for y
m2 = rfr(n_estimators=100,n_jobs=-1,oob_score=True) # create model
%time m2.fit(X_train,y_train) # train the model
print_score(m2,X_train,y_train,X_test,y_test) # see how well you did
```
